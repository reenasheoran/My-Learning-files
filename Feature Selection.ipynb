#### Type 2: Feature selection With Correlation
we will remove features that are highly correlated

```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=load_boston()
data
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df
```

```python
df['TARGET']=data.target
```

```python
df.head()
```

```python
X=df.drop('TARGET',axis=1)
Y=df.TARGET
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
X_train.shape,X_test.shape
```

```python
X_train.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(12,10))
cor=X_train.corr()
sns.heatmap(cor,annot=True,cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
# selecting highly correlated features
def high_cor(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold: # we are interested in absolute coefficient value
                col_name=corr_matrix.columns[i]
                col_corr.add(col_name)
    return col_corr  
```

```python
# threshold is normally selected as 0.85 in real world
corr_features=high_cor(X_train,0.7)
len(set(corr_features))
```

```python
set(corr_features)
```

```python
corr_features
```

```python
X_train.drop(corr_features,axis=1)
X_test.drop(corr_features,axis=1)
```

#### Type 3: Feature selection by information gain- mutual information in classification
Mutual information estimates mutual information for discrete target variable. It is a non-negative value, which measures the dependency between the variables. it is equal to 0 when two variables are independent and higher the value higher the dependency. It is based on entropy estiamtion.

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_wine
data=load_wine()
df=pd.DataFrame(data.data,columns=data.feature_names)
df.head()
```

```python
df['WINE']=data.target
df
```

```python
df.WINE.unique()
```

```python
df.info()
```

```python
# train-test-split to avoid overfitting
X=df.drop(['WINE'],axis=1)
Y=df['WINE']
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
X_train,Y_train
```

```python
# determine the mutual information
from sklearn.feature_selection import mutual_info_classif
MI=mutual_info_classif(X_train,Y_train)
MI
```

```python
MI_series=pd.Series(MI)
MI_series.index=X_train.columns
MI_final=MI_series.sort_values(ascending=False)
```

```python
MI_final.plot.bar(figsize=(20,8))
```

##### Another method with mutual info

```python
# selecting top 5 feature,though u can select the number of featuers based on problem statement
from sklearn.feature_selection import SelectKBest
Top5 = SelectKBest(mutual_info_classif,k=5)
Top5.fit(X_train,Y_train)
X_train.columns[Top5.get_support()]
```

#### Type 5: Feature selection by chi-square test(Fisher score)
we compute chi-square test between non-negative features and target class. It is used to evaluate categorical features in a classification task.

```python
import seaborn as sns
df=sns.load_dataset('titanic')
df
```

```python
df.info()
```

```python
df1=df[['sex','embarked','who','adult_male','embark_town','alive','alone','pclass','survived']].copy()
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
# performing label encoding
import numpy as np
df1['sex']=np.where(df1['sex']=="male",1,0)
df1['alive']=np.where(df1['alive']=="no",1,0)
df1['alone']=np.where(df1['alone']==True,1,0)
df1['who']=np.where(df1['who']=="man",1,0)
df1['adult_male']=np.where(df1['adult_male']==True,1,0)
ordinal_label={k:i for i,k in enumerate(df1.embarked.unique(),0)}
df1['embarked']=df1['embarked'].map(ordinal_label)
ord_label={k:i for i,k in enumerate(df1['embark_town'].unique(),0)}
df1['embark_town']=df1['embark_town'].map(ord_label)
df1
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df1.drop(['survived'],axis=1),df1['survived'],test_size=0.3,random_state=0)
```

```python
X_train
```

```python
# chi-square test
from sklearn.feature_selection import chi2
ch_model=chi2(X_train,Y_train)# returns 2 values first f-score array(higher the better) second p-value array(lower the better)
ch_model
```

```python
ch_series=pd.Series(ch_model[1])
ch_series.index=X_train.columns
ch_final=ch_series.sort_values(ascending=True)
ch_final
```
