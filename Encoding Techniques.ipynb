# Target guided ordinal encoding

#1. ordering the labels according to the target
#2.Replace the labels by joint probability of being 1 or 0

```python
df1=pd.read_csv("C:/Users/reena/Desktop/python learning files/titanic.csv",usecols=['Cabin','Survived'])
```

```python
df1['Cabin'].fillna('Missing',inplace=True)
```

```python
df1
```

```python
# to extract cabin first letter
df1.Cabin=df1['Cabin'].astype(str).str[0]
df1
```

```python
df1.Cabin.unique()
```

```python
Ordinal_labels=df1.groupby(['Cabin'])['Survived'].mean()
```

```python
Ordinal_labels=Ordinal_labels.sort_values().index
```

```python
Ordinal_labels
```

```python
ordinal_dict={k:i for i,k in enumerate(Ordinal_labels,0)}
```

```python
ordinal_dict
```

```python
df1['labels']=df1['Cabin'].map(ordinal_dict)
```

```python
df1
```

# Mean Encoding
Mean of values in particular category

```python
mean_labels=df1.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
df1['mean_Cabin']=df1['Cabin'].map(mean_labels)
df1
```

# Probability-Ratio Encoding

```python
Prob_labels=df1.groupby(['Cabin'])['Survived'].mean()
```

```python
Prob_df=pd.DataFrame(Prob_labels)
Prob_df
```

```python
Prob_df['Died']=1-Prob_df.Survived
```

```python
Prob_df
```

```python
Prob_df['Prob_ratio']=Prob_df.Survived/Prob_df.Died
```

```python
Prob_df
```

```python
Prob_map=Prob_df['Prob_ratio'].to_dict()
```

```python
df1['Prob_encode']=df1['Cabin'].map(Prob_map)
```
